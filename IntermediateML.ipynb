```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("melb_data.csv")

y = data.Price

melb_predictors = data.drop(['Price'], axis=1) #Creates a dataset with all the columns except for Price (the target variable)
X = melb_predictors.select_dtypes(exclude=['object']) #Selects only the columns with numerical data (excludes columns with categorical data)

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```
